```python
#dependencies 
import pandas as pd
```

```python
pizza_csv = "pizza.csv"
burger_csv = "hamburgers.csv"
salads_csv = "salad.csv"
sandwiches_csv = "sandwiches.csv"
smoothies_csv = "smoothies.csv"
burritos_csv = "burritos.csv"
```

```python
pizza_df = pd.read_csv(pizza_csv)
burger_df = pd.read_csv(burger_csv)
salads_df = pd.read_csv(salads_csv)
sandwhiches_df = pd.read_csv(sandwiches_csv)
smoothies_df = pd.read_csv(smoothies_csv)
burritos_df = pd.read_csv(burritos_csv)
```

```python
burritos_df.head()
```

```python
#combined pizza and burger DataFrames to be pizza_burger_df
pizza_burger_df = pizza_df.append(burger_df, sort=False)
pizza_burger_df
```

```python
#combined pizza_burger_df and salads DataFrames to be pizzburg_salads_df
pizzburg_salads_df =  pizza_burger_df.append(salads_df, sort=False)
pizzburg_salads_df
```

```python
#combined pizzburg_salads_df and sandwhiches_df DataFrames to be pbss_df
pbss_df = pizzburg_salads_df.append(sandwhiches_df, sort=False)
pbss_df
```

```python
#pbss_df and smoothies_df DataFrames to be pbss_smoothies_df
pbss_smoothies_df = pbss_df.append(smoothies_df, sort=False)
pbss_smoothies_df
```

```python
#pbss_smoothies_df and burritos_df DataFrames to be all_food_df
all_food_df = pbss_smoothies_df.append(burritos_df, sort=False)
all_food_df
```

```python
all_food_df = all_food_df.drop(columns="Unnamed: 0")
all_food_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
all_food_df.to_csv("output/all_food.csv", index=False, header=True)
```
