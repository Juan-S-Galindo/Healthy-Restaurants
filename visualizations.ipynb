```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
#import scipy.stats as st
#from scipy.stats import linregress
import numpy as np

%matplotlib inline
```

```python
Female_Active_csv = "Female Active.csv"
Female_Moderate_csv = "Female Moderate Active.csv"
Female_Sedentary_csv = "Female Sedentary.csv"
Male_Active_csv = "Male Active.csv"
Male_Moderate_csv = "Male Moderate Active.csv"
Male_Sedentary_csv = "Male Sedentary.csv"

Female_Active_df = pd.read_csv(Female_Active_csv)
Female_Moderate_df = pd.read_csv(Female_Moderate_csv)
Female_Sedentary_df = pd.read_csv(Female_Sedentary_csv)
Male_Active_df = pd.read_csv(Male_Active_csv)
Male_Moderate_df = pd.read_csv(Male_Moderate_csv)
Male_Sedentary_df = pd.read_csv(Male_Sedentary_csv)

```

```python
Female_Active_df.rename(columns={'Unnamed: 0': 'Restaurant', '61 =<': 'Active Female 61 or Older', '31 to 60': 'Active Female 31 to 60', '=< 14 to 30': 'Active Female 14 to 30'}, inplace=True)
Female_Active_df = Female_Active_df.set_index('Restaurant')
#located needed restaurants
Female_Active_df = Female_Active_df.loc[["Panera Bread","Smoothie King","Saladworks", "McAlister's Deli","Jamba Juice"]]
Female_Active_df

```

```python
#create a bar plot
FA_bar_plot = Female_Active_df.plot(kind='bar', title="Active Female Nutritional Consumption by Healthy Reataurants", color="pink", align="center")
# assign x and y axis
FA_bar_plot.set_xlabel("Restaurant")
FA_bar_plot.set_ylabel("Active Females")

    
plt.show   
```
