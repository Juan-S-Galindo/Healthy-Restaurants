## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Import data files
combined_data = "gender_df.csv"

# Read the data and the review results
combined_data = pd.read_csv(combined_data)
combined_data.set_index('Restaurant', inplace=True)
combined_data = combined_data.loc[:,['Moderately Active Male 16 to 25', 'Moderately Active Male 15 and 26 to 45', 'Moderately Active Male 14 and 46 to 65', 'Moderately Active Female 19 to 25', 'Moderately Active Female 14 to 18 and 26 to 50', 'Moderately Active Female 51 and Older']]
combined_data = combined_data.rename(columns={"Moderately Active Male 16 to 25": "Male (19-25)", "Moderately Active Male 15 and 26 to 45": "Male (26-45)", "Moderately Active Male 14 and 46 to 65": "Male (46-65)", "Moderately Active Female 19 to 25": "Female (19-25)", "Moderately Active Female 14 to 18 and 26 to 50": "Female (26-50)", "Moderately Active Female 51 and Older": "Female (51+)"})

combined_data.head()
```

## Bar plots

```python
# Generate a bar plot showing number of data points for each age/gender group

#sort by Restaurant
restaurant = combined_data.groupby('Restaurant')

#create bar plot
bar_plot = combined_data.plot(kind='bar', title="All Restaurants by Gender & Age", color=["lightsteelblue", "cornflowerblue", "royalblue", "thistle", "violet", "orchid"])

#assign x, y axis and format plot
bar_plot.set_xlabel('Restaurant')
bar_plot.set_ylabel('Score')
plt.style.use('bmh')
plt.xticks(rotation=75)
plt.ylim([0, 80])

#position legend outside of box
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

#save figure
plt.savefig('All Restaurants.png', bbox_inches='tight')

plt.show()
```
