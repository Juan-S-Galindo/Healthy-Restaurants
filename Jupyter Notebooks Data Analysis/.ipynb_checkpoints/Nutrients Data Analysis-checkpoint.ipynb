```python
import pandas as pd

#Make sure to run the NutrientsConverter in the Jupyter Notebook folder to be able to call the DataFrames Stored. 

#DataFrames with the combine info.
%store -r maleInfo

%store -r femaleInfo

%store -r foodFiltered_df
```

```python
Nutrients = ['Protein (Cal)',
             'Carbohydrate (Cal)', 
             'Fiber (gr)',
             'Total Sugar (gr)', 
             'Total lipid (Cal)', 
             'Total Trans Fat (Cal)',
             'Total Saturated Fat (Cal)',
             'Cholesterol (mg)', 
             'Calcium (mg)',
             'Iron (mg)', 
             'Sodium, (mg)', 
             'Vitamin A (mcg RAE)', 
             'Vitamin C (mg)']

for i in range(len(maleInfo)):
    
    analysis_df = maleInfo[i]
    
    analysis_df['Restaurant Points'] = ""
    
    for index, row in analysis_df.iterrows():
        
        restaurantTally_dict = {}
        
        for food in range(len(foodFiltered_df)):
            
            if foodFiltered_df.loc[food,'Restaurant'] not in restaurantTally_dict.keys():
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] = 0
                
            if (foodFiltered_df.loc[food,'Protein (Cal)'] <= row['Protein (Cal)']) :
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Carbohydrate (Cal)'] <= row['Carbohydrate (Cal)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Fiber (gr)'] <= row['Fiber (gr)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Total Sugar (gr)'] <= row['Total Sugar (gr)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Total Fat (Cal)'] <= row['Total lipid (Cal)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Total Trans Fat (Cal)'] <= row['Total Trans Fat (Cal)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Total Saturated Fat (Cal)'] <= row['Total Saturated Fat (Cal)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Cholesterol (mg)'] <= row['Cholesterol (mg)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Calcium (mg)'] <= row['Calcium (mg)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Iron (mg)'] <= row['Iron (mg)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Sodium (mg)'] <= row['Sodium, (mg)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Vitamin A (mcg RAE)'] <= row['Vitamin A (mcg RAE)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1

            if (foodFiltered_df.loc[food,'Vitamin C (mg)'] <= row['Vitamin C (mg)']):
                restaurantTally_dict[foodFiltered_df.loc[food,'Restaurant']] += 1 
                
        row['Restaurant Points'] = restaurantTally_dict
        
    maleInfo[i] = analysis_df
```
