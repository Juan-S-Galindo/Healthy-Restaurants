```python
import requests
import pandas as pd
from config import app_id, api_key
```

```python
base_url = 'https://trackapi.nutritionix.com/'
austin_lat_long = '30.2672,-97.7431'
austin_northwest_hills = '30.3532,-97.7675'
south_austin = '30.203, -97.8067'
east_austin= '30.2655, -97.7091'
headers = {'x-app-id': app_id, 'x-app-key':api_key}
params = {'ll': east_austin, 'limit': 100, 'distance': '40mi'}
```

```python
full_url = f"{base_url}v2/locations"
full_url
```

```python
response = requests.get(f"{base_url}v2/locations", headers=headers, params=params).json()
```

```python
locations_list = [item for item in response['locations']]
```

```python
austin_east = pd.DataFrame(locations_list)
austin_east.head()
```

```python
austin_east.to_csv('austin_east.csv')
```

```python
austin_central = pd.read_csv('austin_central.csv')
austin_east = pd.read_csv('austin_east.csv')
austin_northwest_hills = pd.read_csv('austin_northwest_hills.csv')
austin_south = pd.read_csv('austin_south.csv')
```

```python
frame = [austin_central, austin_east, austin_northwest_hills, austin_south]
```

```python
result = pd.concat(frame)
```

```python
result.to_csv('austin_restaurants_nutritionix.csv')
```
